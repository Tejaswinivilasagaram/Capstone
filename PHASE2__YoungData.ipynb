```python
#Create a function to generate countplots for categorical variables
plots={1:[111],2:[121,122],3:[131,132,133],4:[221,222,223,224],5:[231,232,233,234,235],6:[231,232,233,234,235,256]}
def countplot(x,y,df):
  rows=int(str(plots[len(y)][0])[0])
  columns=int(str(plots[len(y)][0])[1])
  plt.figure(figsize=(7*columns,7*rows))

  for i,j in enumerate(y):
    plt.subplot(plots[len(y)][i])
    ax=sns.countplot(x=j,hue=x,data=df,palette='Blues',alpha=0.8,linewidth=0.4,edgecolor='black')
    ax.set_title(j)
  return plt.show
```

```python
import numpy as np
from sklearn.datasets import make_classification
from imblearn.over_sampling import RandomOverSampler
from imblearn.under_sampling import RandomUnderSampler

# Generate an imbalanced dataset for demonstration (You should replace this with your dataset)
X, y = make_classification(n_classes=2, weights=[0.9, 0.1], n_samples=1000, random_state=42)

# Oversample the minority class
over_sampler = RandomOverSampler(sampling_strategy='minority', random_state=42)
X_resampled, y_resampled = over_sampler.fit_resample(X, y)


```

```python
import numpy as np

# Count the occurrences of each class in y_resampled
unique, counts = np.unique(y_resampled, return_counts=True)

# Create a dictionary to store the class counts
class_counts = dict(zip(unique, counts))

# Print the class counts
for class_label, count in class_counts.items():
    print(f"Class {class_label}: {count} samples")
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_resampled, y_resampled, test_size=0.5, random_state=101)
```

```python
# Running logistic regression model
from sklearn.linear_model import LogisticRegression
model = LogisticRegression()
model.fit(X_train, y_train)
train_predictions = model.predict(X_train)
test_predictions = model.predict(X_test)
from sklearn import metrics
train_accuracy = accuracy_score(y_train, train_predictions)
test_accuracy = accuracy_score(y_test, test_predictions)
print("training accuracy",train_accuracy)
print("testing accuracy",test_accuracy)
print("")
print(classification_report(y_test,test_predictions))
#print(confusion_matrix(y_test,test_predictions))
conf_matrix = confusion_matrix(y_test,test_predictions)
class_labels = ['Class 0', 'Class 1']
# Plot the confusion matrix as a heatmap
plt.figure(figsize=(8, 6))
sns.heatmap(conf_matrix, annot=True, fmt='d', cmap='Blues',
            xticklabels=class_labels,
            yticklabels=class_labels)
plt.xlabel('Predicted')
plt.ylabel('True')
plt.title('Confusion Matrix')
plt.show()
```

```python
from sklearn.metrics import classification_report

# Assuming you have already trained and made predictions with your model as in your previous code

# Generate the classification report
report = classification_report(y_test, test_predictions, target_names=class_labels, output_dict=True)

# Print the classification report
for label in class_labels:
    print(f"Class: {label}")
    print(f"Precision: {report[label]['precision']:.2f}")
    print(f"Recall: {report[label]['recall']:.2f}")
    print(f"F1-Score: {report[label]['f1-score']:.2f}")
    print(f"Support: {report[label]['support']}\n")

# Calculate and print the overall metrics
print("Overall Precision:", report['accuracy'])
print("Overall Recall:", report['macro avg']['recall'])
print("Overall F1 Score:", report['macro avg']['f1-score'])
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Replace these with your actual algorithm names and accuracy values
#algorithm_names = ["Algorithm 1", "Algorithm 2", "Algorithm 3", "Algorithm 4", "Algorithm 5", "Algorithm 6"]
#accuracy_values = [0.85, 0.92, 0.78, 0.89, 0.95, 0.88]
algorithm_names = ["LR", "RF", "SVM", "ADA", "KNN", "DT"]
accuracy_values = [0.87,0.98,0.90,0.98,0.93,0.93]

# Define different colors for each bar
bar_colors = ['skyblue', 'salmon', 'lightgreen', 'orange', 'lightcoral', 'lightblue']

# Create a figure and axis
fig, ax = plt.subplots(figsize=(10, 6))

# Plotting the bar graph with different colors
bars = ax.bar(algorithm_names, accuracy_values, color=bar_colors)

# Add data values on top of the bars
for bar, value in zip(bars, accuracy_values):
    plt.text(bar.get_x() + bar.get_width() / 2 - 0.15, bar.get_height() + 0.01, f'{value:.2f}', ha='center', va='bottom')

# Customize the plot
ax.set_xlabel('Algorithms')
ax.set_ylabel('Accuracy')
ax.set_title('Accuracy of Different Machine Learning Algorithms')
ax.set_ylim(0.7, 1.0)  # Adjust the y-axis limits if needed

# Show the plot
plt.show()
```
